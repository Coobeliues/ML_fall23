```python
import matplotlib.pyplot as plt
import pandas as pd

#        ( 'data.csv'    )
data = pd.read_csv('upload_DJIA_table.csv')

#       datetime (    )
data['Date'] = pd.to_datetime(data['Date'])

#  
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close'], label='DJIA Close Price', color='b')
plt.xlabel('')
plt.ylabel('  DJIA')
plt.title('   DJIA  ')
plt.legend()
plt.grid(True)
plt.show()
```
