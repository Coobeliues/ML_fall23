```python
import matplotlib.pyplot as plt
import pandas as pd

# Загрузите данные о цене акций и новостях (замените 'data.csv' на ваш файл данных)
data = pd.read_csv('upload_DJIA_table.csv')

# Преобразование столбца с датой в формат datetime (если он не является таковым)
data['Date'] = pd.to_datetime(data['Date'])

# Создайте график
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close'], label='DJIA Close Price', color='b')
plt.xlabel('Дата')
plt.ylabel('Цена акций DJIA')
plt.title('Изменение цены акций DJIA во времени')
plt.legend()
plt.grid(True)
plt.show()
```
