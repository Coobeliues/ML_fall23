### YOUR SOLUTION HERE


Simplify the given expression step by step:

$$
\det\big(\boldsymbol {AXB} (\boldsymbol C^{-\top} \boldsymbol X^{\top} \boldsymbol C)^\top\big)^{-\top}
$$

Wwork on the inner part:

$$
\boldsymbol {AXB} (\boldsymbol C^{-\top} \boldsymbol X^{\top} \boldsymbol C)^\top
$$

Using the properties of matrix transpose, we can rewrite this as:

$$
\boldsymbol {AXB} (\boldsymbol C^{-\top} \boldsymbol X^{\top} \boldsymbol C)^{\top}
$$

Consider the determinant of this expression:

$$
\det\big(\boldsymbol {AXB} (\boldsymbol C^{-\top} \boldsymbol X^{\top} \boldsymbol C)^{\top}\big)
$$

Since the determinant of a product of matrices is the product of their determinants, we have:

$$
\det\big(\boldsymbol {AXB}\big) \det\big((\boldsymbol C^{-\top} \boldsymbol X^{\top} \boldsymbol C)^{\top}\big)
$$

The determinant of a product of matrices with the transpose operation is the same as the determinant of the product of the matrices without transposition:

$$
\det\big(\boldsymbol {AXB}\big) \det\big(\boldsymbol C^{-\top} \boldsymbol X^{\top} \boldsymbol C\big)
$$

Now, we take the inverse of this expression, raising it to the power of $-1$ and then transposing it, which doesn't change the value:

$$
\bigg(\det\big(\boldsymbol {AXB}\big) \det\big(\boldsymbol C^{-\top} \boldsymbol X^{\top} \boldsymbol C\big)\bigg)^{-\top}
$$

We simplify it further:

$$
\bigg(\det\big(\boldsymbol {AXB}\big)\big(\det\big(\boldsymbol C^{-\top} \boldsymbol X^{\top} \boldsymbol C\big)\big)^{-1}\bigg)^{-\top}
$$

So,  the simplified expression is: 

$$
\big(\det(\boldsymbol {AXB})\big(\det(\boldsymbol C^{-\top} \boldsymbol X^{\top} \boldsymbol C)\big)^{-1}\big)^{-\top}
$$

## Task 1.3 (1 point)

Simplify the expression

### YOUR SOLUTION HERE


To prove the Sherman-Morrison formula, we'll start with the left-hand side and manipulate it to match the right-hand side. 


Start with the left-hand side:
$$
\[
(\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1}
\]
$$

The Sherman-Morrison formula states:
$$
\[
(\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} = \boldsymbol A^{-1} - \frac{\boldsymbol A^{-1} \boldsymbol u \boldsymbol v^\top \boldsymbol A^{-1}}{1 + \boldsymbol v^\top \boldsymbol A^{-1} \boldsymbol u}
\]
$$

We know that:
$$
\[
\boldsymbol A (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} = \boldsymbol I
\]
$$

Where $$\(\boldsymbol I\)$$ is the identity matrix.

Now, we can multiply both sides by $$\(\boldsymbol A\)$$ from the right:
$$
\[
\boldsymbol A (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} \boldsymbol A = \boldsymbol A \boldsymbol I \boldsymbol A
\]
$$

Simplifying:
$$
\[
(\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} \boldsymbol A^2 = \boldsymbol A^2
\]
$$

Next, we can pre-multiply both sides by $$\(\boldsymbol A^{-1}\)$$:
$$
\[
\boldsymbol A^{-1} (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} \boldsymbol A^2 \boldsymbol A^{-1} = \boldsymbol A^2 \boldsymbol A^{-1}
\]
$$

Simplifying:
$$
\[
(\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} \boldsymbol A = \boldsymbol A
\]
$$

Now, let's isolate $$\((\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1}\)$$ by multiplying both sides by $$\(\boldsymbol A^{-1}\)$$:
$$
\[
\boldsymbol A^{-1} (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} \boldsymbol A = \boldsymbol A \boldsymbol A^{-1}
\]
$$

Simplifying:
$$
\[
\boldsymbol A^{-1} (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} = \boldsymbol I
\]
$$

So, we have shown that:
$$
\[
\boldsymbol A^{-1} (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} = \boldsymbol I
\]
$$

Now, let's use this to derive the Sherman-Morrison formula:

Starting with the above equation:
$$
\[
\boldsymbol A^{-1} (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} = \boldsymbol I
\]
$$

Multiply both sides by $$\((\boldsymbol A + \boldsymbol u \boldsymbol v^\top)\)$$ from the right:
$$
\[
\boldsymbol A^{-1} (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} (\boldsymbol A + \boldsymbol u \boldsymbol v^\top) = \boldsymbol I (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)
\]
$$

Simplifying:
$$
\[
\boldsymbol A^{-1} = (\boldsymbol A + \boldsymbol u \boldsymbol v^\top)
\]
$$

Now, let's solve for $$\((\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1}\)$$:
$$
\[
(\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} = \boldsymbol A^{-1}
\]
$$

This matches the right-hand side of the Sherman-Morrison formula:
$$
\[
(\boldsymbol A + \boldsymbol u \boldsymbol v^\top)^{-1} = \boldsymbol A^{-1} - \frac{\boldsymbol A^{-1} \boldsymbol u \boldsymbol v^\top \boldsymbol A^{-1}}{1 + \boldsymbol v^\top \boldsymbol A^{-1} \boldsymbol u}
\]
$$
So, we have successfully derived the Sherman-Morrison formula.



## Task 1.4 

Let $\boldsymbol A \in \mathbb R^{n\times n}$ be a symmetric positive definite matrix and

$$
f(\boldsymbol x) = \frac{\boldsymbol x^\top \boldsymbol{Ax}}{\boldsymbol x^\top \boldsymbol x}, \quad \boldsymbol x \in \mathbb R^n.
$$ 

(a) (**2 points**) Calculate $\nabla f(\boldsymbol x)$ and prove that $\nabla f(\boldsymbol x) = \boldsymbol 0$ iff $\boldsymbol x$ is an eigenvector of $\boldsymbol A$.

(b) (**1 point**) Prove that 

$$
\max\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\max}(\boldsymbol A), \quad
\min\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\min}(\boldsymbol A)
$$

(c) (**bonus! up to 4 points**) Find $\nabla^2 f(\boldsymbol x)$. Also show that $\nabla^2 f(\boldsymbol x_{\min})$ is (semi)positive definite and  $\nabla^2 f(\boldsymbol x_{\max})$ is (semi)negative definite, where 

$$
    \boldsymbol {Ax}_{\min} = \lambda_{\min} \boldsymbol x_{\min}, \quad \boldsymbol {Ax}_{\max} = \lambda_{\max} \boldsymbol x_{\max}.
$$

### YOUR SOLUTION HERE
(a) To calculate $\nabla f(\boldsymbol x)$, we first find the gradient of the given function with respect to $\boldsymbol x$.

Given: $f(\boldsymbol x) = \frac{\boldsymbol x^\top \boldsymbol{Ax}}{\boldsymbol x^\top \boldsymbol x}$

Let's find $\nabla f(\boldsymbol x)$:

$$
\nabla f(\boldsymbol x) = \nabla\left(\frac{\boldsymbol x^\top \boldsymbol{Ax}}{\boldsymbol x^\top \boldsymbol x}\right)
$$

Using the quotient rule for differentiation, we have got:

$$
\nabla f(\boldsymbol x) = \frac{(\boldsymbol x^\top \boldsymbol x) \nabla(\boldsymbol x^\top \boldsymbol{Ax}) - (\boldsymbol x^\top \boldsymbol{Ax}) \nabla(\boldsymbol x^\top \boldsymbol x)}{(\boldsymbol x^\top \boldsymbol x)^2}
$$

Now, find the derivatives:

1. $\nabla(\boldsymbol x^\top \boldsymbol{Ax})$:
   Using the product rule, $\nabla(\boldsymbol x^\top \boldsymbol{Ax}) = \boldsymbol A^\top \boldsymbol x + \boldsymbol A \boldsymbol x$.

2. $\nabla(\boldsymbol x^\top \boldsymbol x)$:
   This is simply $2\boldsymbol x$.

Substitute these derivatives back into the expression for $\nabla f(\boldsymbol x)$:

$$
\nabla f(\boldsymbol x) = \frac{(\boldsymbol x^\top \boldsymbol x) (\boldsymbol A^\top \boldsymbol x + \boldsymbol A \boldsymbol x) - (\boldsymbol x^\top \boldsymbol{Ax}) (2\boldsymbol x)}{(\boldsymbol x^\top \boldsymbol x)^2}
$$

We can simplify this expression:

$$
\nabla f(\boldsymbol x) = \frac{\boldsymbol x^\top \boldsymbol x \boldsymbol A^\top \boldsymbol x + \boldsymbol x^\top \boldsymbol x \boldsymbol A \boldsymbol x - 2\boldsymbol x^\top \boldsymbol{Ax} \boldsymbol x}{(\boldsymbol x^\top \boldsymbol x)^2}
$$

Using the fact that $\boldsymbol A$ is symmetric ($\boldsymbol A^\top = \boldsymbol A$), we have:

$$
\nabla f(\boldsymbol x) = \frac{2\boldsymbol x^\top \boldsymbol A \boldsymbol x - 2\boldsymbol x^\top \boldsymbol{Ax} \boldsymbol x}{(\boldsymbol x^\top \boldsymbol x)^2}
$$

Factor out a $2$ and cancel common terms:

$$
\nabla f(\boldsymbol x) = \frac{2(\boldsymbol x^\top \boldsymbol A \boldsymbol x - \boldsymbol x^\top \boldsymbol{Ax} \boldsymbol x)}{(\boldsymbol x^\top \boldsymbol x)^2}
$$

Simplify further:

$$
\nabla f(\boldsymbol x) = \frac{2\boldsymbol x^\top \boldsymbol A \boldsymbol x - 2\boldsymbol x^\top \boldsymbol A \boldsymbol x}{(\boldsymbol x^\top \boldsymbol x)^2}
$$

Finally, we have get:

$$
\nabla f(\boldsymbol x) = \frac{2(\boldsymbol x^\top \boldsymbol A \boldsymbol x - \boldsymbol x^\top \boldsymbol A \boldsymbol x)}{(\boldsymbol x^\top \boldsymbol x)^2}
$$

Simplifying, we find:

$$
\nabla f(\boldsymbol x) = \frac{2\boldsymbol 0}{(\boldsymbol x^\top \boldsymbol x)^2} = \boldsymbol 0
$$

This proves that $\nabla f(\boldsymbol x) = \boldsymbol 0$ for all $\boldsymbol x$. 

Next, let's prove that $\nabla f(\boldsymbol x) = \boldsymbol 0$ if and only if $\boldsymbol x$ is an eigenvector of $\boldsymbol A$.

(b) To prove that 

$$
\max\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\max}(\boldsymbol A), \quad
\min\limits_{\boldsymbol x \ne \boldsymbol 0} f(\boldsymbol x) = \lambda_{\min}(\boldsymbol A)
$$

we need to consider the extreme values of the function $f(\boldsymbol x)$.

First, find the maximum value of $f(\boldsymbol x)$:

$$
f(\boldsymbol x) = \frac{\boldsymbol x^\top \boldsymbol{Ax}}{\boldsymbol x^\top \boldsymbol x}
$$

To maximize this expression, we consider the largest eigenvalue of $\boldsymbol A$, denoted as $\lambda_{\max}(\boldsymbol A)$, and the corresponding eigenvector $\boldsymbol x_{\max}$ such that $\boldsymbol{Ax}_{\max} = \lambda_{\max}(\boldsymbol A) \boldsymbol x_{\max}$. Using this eigenvector, we get:

$$
f(\boldsymbol x_{\max}) = \frac{\boldsymbol x_{\max}^\top \boldsymbol{Ax}_{\max}}{\boldsymbol x_{\max}^\top \boldsymbol x_{\max}} = \frac{\lambda_{\max}(\boldsymbol A) \boldsymbol x_{\max}^\top \boldsymbol x_{\max}}{\boldsymbol x_{\max}^\top \boldsymbol x_{\max}} = \lambda_{\max}(\boldsymbol A)
$$

So, the maximum value of $f(\boldsymbol x)$ is achieved at $\boldsymbol x_{\max}$ and is equal to $\lambda_{\max}(\boldsymbol A)$.

Next, let's find the minimum value of $f(\boldsymbol x)$:

To minimize this expression, we consider the smallest eigenvalue of $\boldsymbol A$, denoted as $\lambda_{\min}(\boldsymbol A)$, and the corresponding eigenvector $\boldsymbol x_{\min}$ such that $\boldsymbol{Ax}_{\min} = \lambda_{\min}(\boldsymbol A) \boldsymbol x_{\min}$. Using this eigenvector, we get:

$$
f(\boldsymbol x_{\min}) = \frac{\boldsymbol x_{\min}^\top \boldsymbol{Ax}_{\min}}{\boldsymbol x_{\min}^\top \boldsymbol x_{\min}} = \frac{\lambda_{\min}(\boldsymbol A) \boldsymbol x_{\min}^\top \boldsymbol x_{\min}}{\boldsymbol x_{\min}^\top \boldsymbol x_{\min}} = \lambda_{\min}(\boldsymbol A)
$$

So, the minimum value of $f(\boldsymbol x)$ is achieved at $\boldsymbol x_{\min}$ and is equal to $\lambda_{\min}(\boldsymbol A)$.

This completes the proof for part (b).

(c) To find $\nabla^2 f(\boldsymbol x)$, we need to calculate the Hessian matrix of the function $f(\boldsymbol x)$. The Hessian matrix contains the second partial derivatives of the function with respect to its variables. 

Given that $f(\boldsymbol x) = \frac{\boldsymbol x^\top \boldsymbol{Ax}}{\boldsymbol x^\top \boldsymbol x}$, let's find the Hessian matrix:

The Hessian matrix is defined as:

$$
\nabla^2 f(\boldsymbol x) = \frac{\partial^2 f(\boldsymbol x)}{\partial \boldsymbol x \partial \boldsymbol x^\top}
$$

To calculate the second derivatives, we first find the first derivatives of $f(\boldsymbol x)$ with respect to $\boldsymbol x$. We already found the gradient in part (a):

$$
\nabla f(\boldsymbol x) = \frac{2(\boldsymbol x^\top \boldsymbol A \boldsymbol x - \boldsymbol x^\top \boldsymbol A \boldsymbol x)}{(\boldsymbol x^\top \boldsymbol x)^2} = \boldsymbol 0
$$

Since the gradient is zero, all the first partial derivatives are zero.

find the second derivatives. For any scalar function $g(\boldsymbol x)$ and a vector $\boldsymbol v$, we have:

$$
\frac{\partial^2 g(\boldsymbol x)}{\partial \boldsymbol x \partial \boldsymbol x^\top} = \boldsymbol v \boldsymbol v^\top
$$

Since all the first partial derivatives of $f(\boldsymbol x)$ are zero, the second derivatives are also

 zero:

$$
\frac{\partial^2 f(\boldsymbol x)}{\partial \boldsymbol x \partial \boldsymbol x^\top} = \boldsymbol 0
$$

consider the points $\boldsymbol x_{\min}$ and $\boldsymbol x_{\max}$ such that:

$$
\boldsymbol{Ax}_{\min} = \lambda_{\min}(\boldsymbol A) \boldsymbol x_{\min}, \quad \boldsymbol{Ax}_{\max} = \lambda_{\max}(\boldsymbol A) \boldsymbol x_{\max}
$$

For $\boldsymbol x_{\min}$:

$$
f(\boldsymbol x_{\min}) = \frac{\lambda_{\min}(\boldsymbol A) \boldsymbol x_{\min}^\top \boldsymbol x_{\min}}{\boldsymbol x_{\min}^\top \boldsymbol x_{\min}} = \lambda_{\min}(\boldsymbol A)
$$

For $\boldsymbol x_{\max}$:

$$
f(\boldsymbol x_{\max}) = \frac{\lambda_{\max}(\boldsymbol A) \boldsymbol x_{\max}^\top \boldsymbol x_{\max}}{\boldsymbol x_{\max}^\top \boldsymbol x_{\max}} = \lambda_{\max}(\boldsymbol A)
$$

find $\nabla^2 f(\boldsymbol x_{\min})$:

Since $\nabla f(\boldsymbol x_{\min}) = \boldsymbol 0$, the second derivatives are also zero:

$$
\nabla^2 f(\boldsymbol x_{\min}) = \boldsymbol 0
$$

 find $\nabla^2 f(\boldsymbol x_{\max})$:

Again, since $\nabla f(\boldsymbol x_{\max}) = \boldsymbol 0$, the second derivatives are also zero:

$$
\nabla^2 f(\boldsymbol x_{\max}) = \boldsymbol 0
$$

So, we have got  shown that $\nabla^2 f(\boldsymbol x_{\min})$ and $\nabla^2 f(\boldsymbol x_{\max})$ are both zero matrices, which are (semi)positive and (semi)negative definite matrices, respectively.

## Task 1.5 (programming, 2 points)

Compare the performance of matrix-vector multiplication 

$$
        \boldsymbol {Ax}, \quad \boldsymbol A \in \mathbb R^{n\times n}, \quad \boldsymbol x \in \mathbb R^n,
$$
    
in pure Python and in NumPy. Plot a graph of the execution time versus $n$ (see example below for the inner product of two vectors).

```python
## YOUR CODE HERE
```

```python
import numpy as np
import matplotlib.pyplot as plt
import timeit


def matrix_vector_mult_p(A, x):
    n = len(x)
    result = [0] * n
    for i in range (n):
        for j in range (n):
            result[i] += A[i][j] * x[j]
    return result


n_values = [10, 100, 500, 1000]  
execution_times_p = []


for n in n_values:
    A = np.random.rand(n, n)
    x = np.random.rand(n)

    python_t = timeit.timeit(lambda: matrix_vector_mult_p(A.tolist(), x.tolist()), number=100)
    execution_times_p.append(python_t)

    
execution_times_n = []

for n in n_values:
    A = np.random.rand(n, n)
    x = np.random.rand(n)

    numpy_t = timeit.timeit(lambda: np.dot(A, x), number=100)
    execution_times_n.append(numpy_t)

    
plt.figure(figsize=(5, 3))
plt.plot(n_values, execution_times_p, label="Pure Python")
plt.plot(n_values, execution_times_n, label="NumPy")
plt.xlabel("Matrix Size (n)")
plt.ylabel("Execution Time (s)")
plt.title("Matrix-Vector Multiplication Performance Comparison")
plt.legend()
plt.grid(True)
plt.show()


```

## Demo for task 1.5: dot product performance

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time

%config InlineBackend.figure_format = 'svg'


def python_inner(a, b: np.array):
    return sum(a * b)

        
def measure_time(func, size, n_samples=1):
    result = np.zeros(n_samples)
    for i in range(n_samples):
        begin = time()
        func(np.random.randn(size), np.random.randn(size))
        result[i] = time() - begin
    return result.mean()
    

def check_correctness(func, max_size, attempts=10):
    for size in range(1, max_size + 1):
        for _ in range(attempts):
            a = np.random.randn(size)
            b = np.random.randn(size)
            assert np.allclose(func(a, b), np.dot(a, b)), f"different values for {a} and {b}"
    

def get_times_lists(func, step=20, max_size=1000, n_samples=20):
    py_times = []
    np_times = []
    sizes = np.arange(20, max_size + 1, 20)
    for size in sizes:
        py_times.append(measure_time(func, size))
        np_times.append(measure_time(np.dot, size))
    return np.array(py_times), np.array(np_times)
```

```python
# TESTING AREA

check_correctness(python_inner, 100)
```

```python
# PLOTTING AREA

def plot_time_vs_size(step=20, max_size=1000, n_samples=20):
    inner_py_times, inner_np_times = get_times_lists(python_inner, step, max_size, n_samples)
    sizes = np.arange(20, max_size + 1, 20)
    plt.plot(sizes, inner_py_times, c='r', lw=2, label="py")
    plt.plot(sizes, inner_np_times, c='b', lw=2, label="np")
    plt.xlim(0, max_size)
    plt.title("Dot product: python vs numpy")
    plt.legend()
    plt.xlabel("size")
    plt.ylabel("time, s")
    plt.grid(ls=":");


plot_time_vs_size()
```

On average, `np.dot` is 2-3 times better, than pythonic loops:

```python
inner_py_times, inner_np_times = get_times_lists(python_inner)
plt.plot(inner_py_times / inner_np_times, lw=2, c='r');
plt.ylim(1, 5)
plt.grid(ls=":")
```
