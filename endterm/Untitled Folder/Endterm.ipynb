```python
df = pd.read_csv("data.csv")
def make_proper_student_index(dff):
    student_ques = {
        '30': 0.5,
        '29': 1,
        '27': [0, 0.25, 0.5],
        '26': 0.25,
        '25': [-0.25, 0.25, 0.5],
        '24': [0.25, 0.5, -0.25],
        '23': [1, 0.5, 0],
        '22': [0.5, 0.25, -0.25],
        '20': [0.25, 0],
        '17': 0.25
    }
    dff["PS index"] = 0
    for key, value in student_ques.items():
        if(type(value) != type([])):
            dff["PS index"] += dff[key] * value
        else:
            dff["PS index"] += pd.Series([value[val - 1] for val in list(dff[key])])
    dff.drop(list(student_ques.keys()), axis = 1, inplace = True)
    return dff
def make_proper_family_index(dff):
    student_ques = {
        '2': 0.1,
        '3': [0, 0.1,-0.1],
        '5': [-0.1, 0.1],
        '8': -0.05,
        '11': [-0.1, -0.1, -0.1, 0, 0, 0],
        "13": [0, -0.1, 0, -0.1, 0.05, 0],
        "15": [0, -0.1, 0, 0, 0, 0]
    }
    dff["F index"] = 0
    for key, value in student_ques.items():
        if(type(value) != type([])):
            dff["F index"] += dff[key] * value
        else:
            dff["F index"] += pd.Series([value[val - 1] for val in list(dff[key])])
    dff.drop(list(student_ques.keys()), axis = 1, inplace = True)
    return dff
df = make_proper_student_index(df)
df = make_proper_family_index(df)
```

<h7> 1) Building Stratified Cross Validation Function. </h7>
 


```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.base import clone
X = df.drop("GRADE", axis = 1); y = df["GRADE"]
def stratified_cross_val(model):
    skf = StratifiedShuffleSplit(n_splits = 5, test_size = 0.25)
    skf.get_n_splits(X, y)
    scores = []
    for i, (train_index, test_index) in enumerate(skf.split(X, y)):
        model_clone = clone(model)
        model_clone.fit(X.iloc[train_index, :], y[train_index])
        predict = model_clone.predict(X.iloc[test_index, :])
        scores.append(np.sum(predict == y[test_index])/len(test_index))
    return scores
```

`It's important to preserve balance in distributions of training and test sets, since amount of instances isn't enough to guarantee unbiased training using train_test_split, considering number of possible classes.` <br>
`I will demonstrate reasonability of this statement by showing one example using train_test_split function.`

```python
from sklearn.model_selection import GridSearchCV
X = df.drop("GRADE", axis = 1); y = df["GRADE"]
```

```python
des_tree = DecisionTreeClassifier(random_state = 42)
ranforest = RandomForestClassifier(random_state = 42)
sgd = SGDClassifier(random_state = 42)
lr = LogisticRegression(max_iter=1000, random_state = 42)
svc = SVC(random_state = 42, probability = True)

def gridsearchmodels(metric):
    param_grid = {"splitter": ['best', 'random'], 
                  'criterion' : ['gini', 'log_loss', 'entropy'],  
                  'max_depth' : list(range(1, 10)) + [None],
                  'max_leaf_nodes': list(range(2, 10)) + [None],
                  'min_samples_split' : list(range(2, 6)),
                  'min_samples_leaf' : list(range(1, 6)),
                 }
    des_tree_gs = GridSearchCV(des_tree, param_grid, cv = StratifiedShuffleSplit(n_splits = 5, test_size = 0.25), scoring = metric, n_jobs = -1)
    des_tree_gs.fit(X, y)
    param_grid = {"n_estimators": [100], 
              'max_depth' : list(range(2, 10)) + [None],
              'max_leaf_nodes': list(range(2, 8)) + [None],
              'min_samples_split' : list(range(2, 6)),
              'min_samples_leaf' : list(range(1, 6)),
              'bootstrap': [True]}
    ranforest_gs = GridSearchCV(ranforest, param_grid, cv = StratifiedShuffleSplit(n_splits = 5, test_size = 0.25), scoring = metric, n_jobs = -1)
    ranforest_gs.fit(X, y)
    param_grid = {'loss': ['hinge', 'log_loss', 'modified_huber', 'squared_hinge', 'perceptron', 'squared_error', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive'],
              'penalty': ['l2', 'l1', 'elasticnet', None],
              'alpha': list(np.linspace(0.0001, 5, 20)),
              'l1_ratio': list(np.linspace(0, 1, 20)),
              'shuffle': [True, False]}
    sgd_gs = GridSearchCV(sgd, param_grid, cv = StratifiedShuffleSplit(n_splits = 5, test_size = 0.25), scoring = metric, n_jobs = -1)
    sgd_gs.fit(X, y)
    param_grid = {'penalty': ['l2', 'l1', 'elasticnet', None],
              'C': list(np.linspace(0.001, 5, 20))}
    lr_gs = GridSearchCV(lr, param_grid, cv = StratifiedShuffleSplit(n_splits = 5, test_size = 0.25), scoring = metric, n_jobs = -1)
    lr_gs.fit(X, y)
    param_grid = {'kernel' : ['linear','rbf', 'poly'],
              'degree': [1, 2, 3, 4],
              'C': list(np.linspace(0.001, 5, 20)),
              'gamma': ['auto', 'scale']
             }
    svc_gs = GridSearchCV(svc, param_grid, cv = StratifiedShuffleSplit(n_splits = 5, test_size = 0.25), scoring = metric, n_jobs = -1)
    svc_gs.fit(X, y)
    return [lr_gs, sgd_gs, des_tree_gs, ranforest_gs, svc_gs]
```
